```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../data/alcaldes.csv', sep= '|')
```

```python
ax = df.groupby(['Partido']).size()\
.sort_values(ascending=False)[:26]\
.plot.bar(title = u'Movimientos Ganadores', legend=True, figsize=(16,5))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    # ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
```
